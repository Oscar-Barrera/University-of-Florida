```python
import pandas as pd
gbms = pd.DataFrame(trigcat.get_table(columns=('version', 'trigger_name', 'name', 'ra', 'dec', 'trigger_time',
       'trigger_type', 'reliability', 'adc_high', 'adc_low', 'bii',
       'channel_high', 'channel_low', 'dec_scx', 'dec_scz',
       'detector_mask', 'end_time', 'error_radius', 'geo_lat', 'geo_long',
       'lii', 'localization_source', 'phi', 'ra_scx', 'ra_scz', 'theta',
       'time', 'trigger_algorithm', 'trigger_timescale')))

gbms
```

```python
#get trigger number only

gbms['trigger'] = gbms.trigger_name.str[2:-3]

gbms
```

```python
#get triggers we're interested in:

import os

trigs = os.listdir('Spectrum Plots')
trigs = [trig[3:-4] for trig in trigs if '.ipynb' not in trig]

#find where df includes trigs

singles = gbms.loc[gbms['trigger'].isin(trigs)].reset_index(drop=True)
singles = singles.drop(columns = ['trigger'])

singles

# len(set(singles['trigger']))
```

```python
#import All-Wise data

allwise = pd.read_csv('./allwise.txt', sep = "|")
allwise = allwise.loc[:, ~allwise.columns.str.contains('^Unnamed')]
allwise.columns = ['name', 'ra', 'dec']
allwise
```

```python
#create query list

#format: ra, dec; ra2, dec2; ...

coords = []

for index, val in enumerate(singles['ra']):
    coords.append(str(round(val,4)) + ", " + str(round(singles['dec'][index],4)) + "; " )

coords = set(coords)

query = "".join(coords)

query
```
