```python
data_mass2 = pd.read_excel (r'/Users/Oscar/Documents/Research/GW190521.xlsx'
                           ,sheet_name="M11" , engine='openpyxl')

df2 = pd.DataFrame(data_mass2)
```

```python
ax1 = sns.histplot(data =df2, x = 'M11', element = 'poly', 
                  weights = 'M11_Weight', stat = 'density',linewidth = 2.5,
                 bins = 20, fill = False, common_norm = False, color = 'C0')

sns.histplot(data =df2, x = 'M11', element = 'poly', 
                  weights = 'PLWeight_M11', stat = 'density',linewidth = 2.5,
                 bins = 20, fill = False, common_norm = False, color = 'C1')

sns.histplot(data =df2, x = 'M11', element = 'poly', 
                  weights = 'AGNWeight_M11', stat = 'density',linewidth = 2.5,
                 bins = 12, fill = False, common_norm = False, color = 'C2')

ax1.vlines(50, 0, 0.04, linestyles = 'dashed', colors = 'grey')
ax1.set_ylim([0,0.04])

plt.rc('legend', fontsize=14, loc = 'upper right') 
leg = ax1.legend(labels=["Unweighted", "Power Law", "AGN"],
                 fancybox=True, shadow = True, bbox_to_anchor=(0.95, .95))
leg.legendHandles[0].set_color('C0')
leg.legendHandles[1].set_color('C1')
leg.legendHandles[2].set_color('C2')

ax1.set_title('M11', fontsize = 14, fontweight = "bold")
ax1.set_xlabel('Mass [$M_\odot$]', fontsize = 14)
ax1.set_ylabel('PDF', fontsize = 14)
ax1.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```
