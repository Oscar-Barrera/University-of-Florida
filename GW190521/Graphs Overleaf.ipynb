```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
m1 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m1distribution.csv")
m2 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m2distribution.csv")

m11 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m11distribution.csv")
m12 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m12distribution.csv")
m21 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m21distribution.csv")
m22 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m22distribution.csv")

m111 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m111distribution.csv")
m112 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m112distribution.csv")
m121 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m121distribution.csv")
m122 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m122distribution.csv")
m211 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m211distribution.csv")
m212 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m212distribution.csv")
m221 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m221distribution.csv")
m222 = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/m222distribution.csv")

astropri = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/primaryastrophysical.csv")
astrosec = pd.read_csv("/Users/Oscar/Documents/Research/GW190521/Overleaf Distributions/secondaryastrophysical.csv")
```

```python
#m1 and m2 distributions
import h5py

fn = '/Users/Oscar/Documents/Research/GW190521/GW190521.h5'
data = h5py.File(fn,'r')
waveform = 'C01:NRSur7dq4'
posterior_samples = data[waveform]['posterior_samples']
samples = np.array(posterior_samples).T

mass_1=posterior_samples['mass_1_source']
hist_mass_1=np.histogram(mass_1,bins=100,density=True)
dx=hist_mass_1[1][1]-hist_mass_1[1][0]
m1array = (hist_mass_1[1][:-1]+dx,hist_mass_1[0])

mass_2=posterior_samples['mass_2_source']
hist_mass_2=np.histogram(mass_2,bins=100,density=True)
dx=hist_mass_2[1][1]-hist_mass_2[1][0]
m2array = (hist_mass_2[1][:-1]+dx,hist_mass_2[0])
```

```python
plt.rcParams['figure.figsize'] = [14, 5]

fig, axs = plt.subplots(ncols = 2, sharey = True)


#Third Generation
sns.histplot(data = m111, x = 'primary', element = 'poly', weights = 'probabilitym111', stat = 'density',
                  linewidth = 2.5, bins = 67, fill = False, common_norm = False, color = 'C0', ax=axs[0])
sns.histplot(data = m112, x = 'secondary', element = 'poly', weights = 'probabilitym112', stat = 'density',
                  linewidth = 2.5, bins = 15, fill = False, common_norm = False, color = 'C0', ax=axs[0])  
sns.histplot(data = m121, x = 'primary', element = 'poly', weights = 'probabilitym121', stat = 'density',
                  linewidth = 2.5, bins = 9, fill = False, common_norm = False, color = 'C0', ax=axs[0])
sns.histplot(data = m122, x = 'secondary', element = 'poly', weights = 'probabilitym122', stat = 'density',
                  linewidth = 2.5, bins = 27, fill = False, common_norm = False, color = 'C0', ax=axs[0])
#----------------
sns.histplot(data = m211, x = 'primary', element = 'poly', weights = 'probabilitym211', stat = 'density',
                  linewidth = 2.5, bins = 77, fill = False, common_norm = False, color = 'C0', ax=axs[1])
sns.histplot(data = m212, x = 'secondary', element = 'poly', weights = 'probabilitym212', stat = 'density',
                  linewidth = 2.5, bins = 40, fill = False, common_norm = False, color = 'C0', ax=axs[1])
sns.histplot(data = m221, x = 'primary', element = 'poly', weights = 'probabilitym221', stat = 'density',
                  linewidth = 2.5, bins = 7, fill = False, common_norm = False, color = 'C0', ax=axs[1])
sns.histplot(data = m222, x = 'secondary', element = 'poly', weights = 'probabilitym222', stat = 'density',
                  linewidth = 2.5, bins = 24, fill = False, common_norm = False, color = 'C0', ax=axs[1])


#Second Generation
sns.histplot(data = m11, x = 'm11_mass', element = 'poly', weights = 'probabilitym11', stat = 'density',
                  linewidth = 2.5, bins = 30, fill = False, common_norm = False, color = 'C1', ax=axs[0])
sns.histplot(data = m12, x = 'm12_mass', element = 'poly', weights = 'probabilitym12', stat = 'density',
                  linewidth = 2.5, bins = 28, fill = False, common_norm = False, color = 'C1', ax=axs[0])
#----------------
sns.histplot(data = m21, x = 'm21_mass', element = 'poly', weights = 'probabilitym21', stat = 'density',
                  linewidth = 2.5, bins = 78, fill = False, common_norm = False, color = 'C1', ax=axs[1])
sns.histplot(data = m22, x = 'm22_mass', element = 'poly', weights = 'probabilitym22', stat = 'density',
                  linewidth = 2.5, bins = 18, fill = False, common_norm = False, color = 'C1', ax=axs[1])



#First Generation

ax1 = sns.histplot(x = m1array[0], element = 'poly', weights = m1array[1], stat = 'density',
                  linewidth = 2.5, bins = 100, fill = False, common_norm = False, color = 'C3', ax=axs[0])
#----------------
ax2 = sns.histplot(x = m2array[0], element = 'poly', weights = m2array[1], stat = 'density',
                  linewidth = 2.5, bins = 98, fill = False, common_norm = False, color = 'C3', ax=axs[1])


#Astrophysical
sns.histplot(data = astropri, x = 'piastroprimary_mass', element = 'poly', weights = 'astroprimary', 
             stat = 'density', linewidth = 2.5, bins = 18, fill = False, common_norm = False, 
             color = 'k', ax=axs[0], linestyle = '--')
sns.histplot(data = astrosec, x = 'piastrosecondary_mass', element = 'poly', weights = 'astrosecondary', 
             stat = 'density', linewidth = 2.5, bins = 39, fill = False, common_norm = False, 
             color = 'k', ax=axs[1], linestyle = '--')

#Set grey area between 0 and 5
ax1.plot(range(1))
ax1.axvspan(0, 5, alpha=0.3, color='grey')
ax2.plot(range(1))
ax2.axvspan(0, 5, alpha=0.3, color='grey')

#verticle line at 50 solar masses
ax2.vlines(50, 0, 1, linestyles = 'solid', colors = 'grey')
ax1.vlines(50, 0, 1, linestyles = 'solid', colors = 'grey')

#Graph Limits
axs[0].set_xlim([0,119.9])
axs[1].set_xlim([0,120])
axs[0].set_ylim([0,.085])
axs[1].set_ylim([0,.085])

#Legend
plt.rc('legend', fontsize=14, loc = 'upper right') 
leg1 = ax1.legend(labels=["$\it{M_{1}}$", r"$\it{Parents}$", r"$\it{Grandparents}$",r"$\it{GWTC-2}$"],
                 fancybox=True, shadow = True, bbox_to_anchor=(1, 1))
leg2 = ax2.legend(labels=["$\it{M_{2}}$", r"$\it{Parents}$", r"$\it{Grandparents}$",r"$\it{GWTC-2}$"],
                 fancybox=True, shadow = True, bbox_to_anchor=(1, 1))

#Set legend styles
for line in leg1.get_lines():
    line.set_linewidth(3.0)
for line in leg2.get_lines():
    line.set_linewidth(3.0)
    
leg1.legendHandles[0].set_color('C3')
leg1.legendHandles[1].set_color('C1')
leg1.legendHandles[2].set_color('C0')
leg1.legendHandles[3].set_color('k')

leg1.legendHandles[3].set_linestyle('--')

leg2.legendHandles[0].set_color('C3')
leg2.legendHandles[1].set_color('C1')
leg2.legendHandles[2].set_color('C0')
leg2.legendHandles[3].set_color('k')

leg2.legendHandles[3].set_linestyle('--')

#Grid
axs[0].grid(linestyle='--', linewidth=1)
axs[1].grid(linestyle='--', linewidth=1)

#Titles and Labels
ax1.set_title('GW190521 Mass 1', fontsize = 14, fontweight = "bold")
ax2.set_title('GW190521 Mass 2', fontsize = 14, fontweight = "bold")
axs[0].set_xlabel('Mass [$M_\odot$]', fontsize = 14)
axs[1].set_xlabel('Mass [$M_\odot$]', fontsize = 14)
axs[0].set_ylabel('PDF', fontsize = 14)

ax1.tick_params(axis='both', which='major', labelsize=14)
ax2.tick_params(axis='both', which='major', labelsize=14)

#Join Graphs Together
plt.subplots_adjust(wspace = 0.05)

#plt.savefig('/Users/Oscar/Documents/Research/GW190521_GWTC_Distribution.pdf')

plt.show()


```
